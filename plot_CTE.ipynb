```python
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
steer = list()
cte = list()


#with open('sample-laser-radar-measurement-data-1.txt') as f:
with open('outfile.txt') as f:
    line = f.readline()
    line = f.readline()
    while line != "":
        print(line)
        row = line.split('\n')[0].split('\t')
#         print(row)
        if (len(row)==2):
            steer.append(row[0])
            cte.append(row[1])
        
        line = f.readline()
```
