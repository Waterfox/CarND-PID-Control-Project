```python
import pandas as pd
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
x = list()
y = list()
vx = list()
vy = list()

x_m = list()
y_m = list()

x_gt = list()
y_gt = list()
vx_gt = list()
vy_gt = list()

#with open('sample-laser-radar-measurement-data-1.txt') as f:
with open('dataset1.txt') as f:
    line = f.readline()
    while line != "":
        row = line.split('\n')[0].split('\t')
#         print(row)
        x.append(row[0])
        y.append(row[1])
        vx.append(row[2])
        vy.append(row[3])
        
        x_m.append(row[4])
        y_m.append(row[5])
        
        x_gt.append(row[6])
        y_gt.append(row[7])
        vx_gt.append(row[8])
        vy_gt.append(row[9])
        
        line = f.readline()
```

```python
fig = plt.figure(figsize=(10,10))
fig.suptitle("Kalman Filter output: Blue: Filter, Red: Measurements, Green: GT",fontsize=20)
ax1 = plt.scatter(x,y,marker='.')
plt.axis([4,8,-14,-10])
ax11 = plt.plot(x,y,linewidth=0.3)
ax2 = plt.scatter(x_m,y_m,c='r',marker='*')
ax3 = plt.scatter(x_gt,y_gt,c='g',marker='v')
ax33 = plt.plot(x_gt,y_gt,c='g',linewidth=0.3)
# plt.scatter(x,y)
```
